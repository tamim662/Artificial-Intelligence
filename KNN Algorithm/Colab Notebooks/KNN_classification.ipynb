```
from numpy import genfromtxt

data_path = '/content/gdrive/My Drive/AI /Dataset/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```
data = my_data.tolist()
for i in range(5):
  print(data[i])
type(data)
```

```
import random
train_set=[]
val_set=[]
test_set=[]
random.shuffle(data)

```

```
#Split data set
for s in data:
  r=random.uniform(0,1)
  if r>=0 and r<=0.7 and len(train_set)/len(data)<=0.7 :
    train_set.append(s)
  elif r>0.7 and r<=0.85 and len(val_set)/len(data)<=0.15 :
    val_set.append(s)
  elif r >0.85 and r<=1 and len(test_set)/len(data) <= 0.15:
    test_set.append(s)

print(len(train_set)/len(data))
print(len(val_set)/len(data))
print(len(test_set)/len(data))

```

```
# Validation Check
import math
k=5
cnt=0
for v in val_set:
  l=[]
  for t in train_set:
    dist=math.sqrt((v[0]-t[0])**2+(v[1]-t[1])**2+(v[2]-t[2])**2+(v[3]-t[3])**2)
    l.append([dist,t])
  l.sort()
  l=l[:k]
  
  frequency=[]
  for i in range(k):
    frequency.append(l[i][1][-1])
  cls=max(frequency,key=frequency.count)
  if v[-1]==cls:
    cnt+=1
    

accuracy=(cnt/len(val_set))*100
print(accuracy)


```

```
#Test case Check
k=10
cnt=0
for v in test_set:
  l=[]
  for t in train_set:
    dist=math.sqrt((v[0]-t[0])**2+(v[1]-t[1])**2+(v[2]-t[2])**2+(v[3]-t[3])**2)
    l.append([dist,t])
  l.sort()
  l=l[:k]
  
  frequency=[]
  for i in range(k):
    frequency.append(l[i][1][-1])
  cls=max(frequency,key=frequency.count)
  if v[-1]==cls:
    cnt+=1
    

accuracy=(cnt/len(test_set))*100
print(accuracy)
```
