```
from numpy import genfromtxt

data_path = '/content/gdrive/My Drive/AI /Dataset/diabetes.csv'
my_data = genfromtxt(data_path, delimiter=',')
my_data
```

```
data = my_data.tolist()
for i in range(5):
  print(data[i])
type(data)
```

```
import random
train_set=[]
val_set=[]
test_set=[]
random.shuffle(data)
```

```
#Split data set
for s in data:
  r=random.uniform(0,1)
  if r>=0 and r<=0.7 and len(train_set)/len(data)<=0.7 :
    train_set.append(s)
  elif r>0.7 and r<=0.85 and len(val_set)/len(data)<=0.15 :
    val_set.append(s)
  elif r >0.85 and r<=1 and len(test_set)/len(data) <= 0.15:
    test_set.append(s)

print(len(train_set)/len(data))
print(len(val_set)/len(data))
print(len(test_set)/len(data))
```

```
# Validation Check
import math
k=15
error=0
for v in val_set:
  l=[]
  for t in train_set:
    dist=math.sqrt((v[0]-t[0])**2+(v[1]-t[1])**2+(v[2]-t[2])**2+(v[3]-t[3])**2+(v[4]-t[4])**2+(v[4]-t[4])**2+(v[5]-t[5])**2+(v[6]-t[6])**2+(v[7]-t[7])**2+(v[8]-t[8])**2+(v[9]-t[9])**2)
    l.append([dist,t])
  l.sort()
  l=l[:k]
  
  frequency=[]
  for i in range(k):
    frequency.append(l[i][1][-1])
  
  avg=sum(frequency)/len(frequency)
  error+=(v[-1]-avg)**2
  

mean_sq_error=error/len(val_set)
    
print(mean_sq_error)
```

```
#Test case Check
k=10
error=0
for v in test_set:
  l=[]
  for t in train_set:
    dist=math.sqrt((v[0]-t[0])**2+(v[1]-t[1])**2+(v[2]-t[2])**2+(v[3]-t[3])**2+(v[4]-t[4])**2+(v[4]-t[4])**2+(v[5]-t[5])**2+(v[6]-t[6])**2+(v[7]-t[7])**2+(v[8]-t[8])**2+(v[9]-t[9])**2)
    l.append([dist,t])
  l.sort()
  l=l[:k]
  
  frequency=[]
  for i in range(k):

    frequency.append(l[i][1][-1])

  avg=sum(frequency)/len(frequency)
  error+=(v[-1]-avg)**2
  

mean_sq_error=error/len(test_set)
    
print(mean_sq_error)
```
